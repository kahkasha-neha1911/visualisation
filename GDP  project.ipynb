```python
import pandas as pd 
df = pd.read_csv('gdp.csv')
df.head()
```

```python
df.isnull().value_counts()
```

```python
df.dtypes
```

```python
df.rename(columns = {'Life expectancy at birth (years)' : 'LEB'} , inplace = True)
df.head()
```

```python
df['Country'].value_counts()
```

```python
import matplotlib.pyplot as plt

import seaborn as sns

Chile = df[df['Country']== 'Chile']

sns.scatterplot(
    data=Chile, x='GDP', y= 'LEB', hue='Year' , palette='viridis',edgecolor='black')

plt.xlabel('Gross Domestic Product (GDP)', fontsize=12)  # X-axis label
plt.ylabel('Life Expectancy at Birth (LEB)', fontsize=12)  # Y-axis label
plt.title('GDP vs. Life Expectancy in Chile Over Years', fontsize=14)  # Title
plt.legend(title='Year')  # Customize legend title
plt.grid(True)
plt.show()
```

```python

China= df[df['Country']== 'China']

sns.scatterplot(
    data= China, x='GDP', y= 'LEB', hue='Year' , palette='coolwarm', edgecolor='black')

plt.xlabel('Gross Domestic Product (GDP)', fontsize=12)  # X-axis label
plt.ylabel('Life Expectancy at Birth (LEB)', fontsize=12)  # Y-axis label
plt.title('GDP vs. Life Expectancy in China Over Years', fontsize=14)  # Title
plt.legend(title='Year')  # Customize legend title
plt.grid(True) 

plt.show()
```

```python

Germany= df[df['Country']== 'Germany']

sns.scatterplot(
    data=Germany, x='GDP', y= 'LEB', hue='Year' ,  palette='magma', edgecolor='black')

plt.xlabel('Gross Domestic Product (GDP)', fontsize=12)  # X-axis label
plt.ylabel('Life Expectancy at Birth (LEB)', fontsize=12)  # Y-axis label
plt.title('GDP vs. Life Expectancy in Germany Over Years', fontsize=14)  # Title
plt.legend(title='Year')  # Customize legend title
plt.grid(True)

plt.show()
```

```python
US = df[df['Country']=='United States of America']
sns.scatterplot(
    data=US, x='GDP', y= 'LEB', hue='Year' , palette='plasma', edgecolor='black')

plt.xlabel('Gross Domestic Product (GDP)', fontsize=12)  # X-axis label
plt.ylabel('Life Expectancy at Birth (LEB)', fontsize=12)  # Y-axis label
plt.title('GDP vs. Life Expectancy in the United States Over Years', fontsize=14)  # Title
plt.legend(title='Year')  # Customize legend title
plt.grid(True)

plt.show()
```

```python
Zimbabwe = df[df['Country']=='Zimbabwe']

sns.scatterplot(
    data=Zimbabwe , x='GDP', y= 'LEB', hue='Year' , palette='coolwarm',edgecolor='black')
plt.xlabel('Gross Domestic Product (GDP)', fontsize=12)  # X-axis label
plt.ylabel('Life Expectancy at Birth (LEB)', fontsize=12)  # Y-axis label
plt.title('GDP vs. Life Expectancy in Zimbabwe Over Years', fontsize=14)  # Title
plt.legend(title='Year')  # Customize legend title
plt.grid(True) 

plt.show()
```

```python
# Create a figure with multiple subplots (1 row, 5 columns)
plt.figure(figsize=(20, 6))
# Plot for Chile (subplot 1)
plt.subplot(1, 5, 1)  # 1 row, 5 columns, position 1
sns.scatterplot(
    data=Chile, x='GDP', y='LEB',  color = 'purple', edgecolor='black'
)
plt.title('Chile', fontsize=14)
plt.xlabel('GDP', fontsize=10)
plt.ylabel('LEB', fontsize=10)


# Plot for China (subplot 2)
plt.subplot(1, 5, 2)  # 1 row, 5 columns, position 2
sns.scatterplot(
    data=China, x='GDP', y='LEB',  color ='blue', edgecolor='black'
)
plt.title('China', fontsize=14)
plt.xlabel('GDP', fontsize=10)
plt.ylabel('LEB', fontsize=10)


# Plot for Germany (subplot 3)
plt.subplot(1, 5, 3)  # 1 row, 5 columns, position 3
sns.scatterplot(
    data=Germany, x='GDP', y='LEB',  color='red', edgecolor='black'
)
plt.title('Germany', fontsize=14)
plt.xlabel('GDP', fontsize=10)
plt.ylabel('LEB', fontsize=10)


# Plot for the United States (subplot 4)
plt.subplot(1, 5, 4)  # 1 row, 5 columns, position 4
sns.scatterplot(
    data=US, x='GDP', y='LEB',  color= 'green', edgecolor='black'
)
plt.title('United States', fontsize=14)
plt.xlabel('GDP', fontsize=10)
plt.ylabel('LEB', fontsize=10)


# Plot for Zimbabwe (subplot 5)
plt.subplot(1, 5, 5)  # 1 row, 5 columns, position 5
sns.scatterplot(
    data=Zimbabwe, x='GDP', y='LEB',  color = 'orange', edgecolor='black'
)
plt.title('Zimbabwe', fontsize=14)
plt.xlabel('GDP', fontsize=10)
plt.ylabel('LEB', fontsize=10)


# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```
